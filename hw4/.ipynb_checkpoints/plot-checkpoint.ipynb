```python
import numpy as np

def sign(x):
    return 1 if x >=0 else -1

def sigmoid(x):
    return 1. / (1 + np.exp(-x))
def dEin(X, Y, w):
    n,d = X.shape
    buf = np.matrix(np.zeros([n,d]))
    for i in range(n):
        x = X[i,:]
        y = float(Y[i])
        s = -y * np.dot(w,x)
        theta = 1.0/(1+np.exp(-s))
        buf[i,:] = theta * (-y*x)
#         buf[i,:] = sigmoid(s) * (-y*x)

        
    gradient_average = np.sum(buf, axis=0) / n
#     return buf
    return gradient_average
def step(X, Y, w, lr):
    return w - lr * dEin(X, Y, w)

def read_data(path):
    with open(path) as f:
        ls = f.readlines()
    n = len(ls)
    #check dimension
    d = len(ls[0].strip().split(' ')) -1
    X = np.zeros([n, d+1])
    Y = np.zeros([n])
    for i,l in enumerate(ls):
        buf = l.strip().split(' ')
        y = float(buf[-1]) 
        Y[i] = y 
        X[i,0] = 1.
        for idx,s in enumerate(buf[:-1]):
            X[i,idx+1] = float(s)
    return X,Y,d

def Ein(X, Y, w):
    err = 0
    for i in range(len(X)):
        x = X[i,:]
        y = Y[i]
        s = float(np.dot(w,x))
        y_ = sign(s)
#         print y_
        err += 1 if y != y_ else 0
    return float(err)



X,Y,d = read_data('./hw3_train.dat.txt')
# X,Y,d = read_data('./t')


test_X,test_Y,_ = read_data('./hw3_test.dat.txt')
# Q18
# w = np.zeros([1, d+1]) 
# lr = 0.001
# for _ in range(2000):
# #     print w
#     w = step(X,Y,w,lr)
    
# eout = Ein(test_X,test_Y,w)/len(test_X)
# print eout

# # Q19
# w = np.zeros([1, d+1]) 
# lr = 0.01
# for _ in range(2000):
# #     print w
#     w = step(X,Y,w,lr)
    
# eout = Ein(test_X,test_Y,w)/len(test_X)
# print eout
# Q20
w = np.zeros([1, d+1]) 
lr = 0.001
for i in range(2000):
    idx = i%n
    X_ = X[idx:idx+1,:]
    Y_ = Y[idx:idx+1]
    w = step(X_,Y_,w,lr)
    
eout = Ein(test_X,test_Y,w)/len(test_X)
print eout





```
